

**Projet : Gestionnaire de Contacts en Python**

**Description du projet** :
Vous allez créer un programme en Python qui vous permettra de gérer une liste de contacts. Chaque contact sera représenté par un dictionnaire contenant des informations telles que le nom, le prénom, le numéro de téléphone, l'adresse e-mail, etc. Vous utiliserez des fonctions de dictionnaire pour ajouter, mettre à jour, supprimer et rechercher des contacts. De plus, vous utiliserez des listes de compréhension pour effectuer des opérations sur les contacts.

**Fonctionnalités du gestionnaire de contacts** :

1. Ajouter un contact : Le programme devrait permettre d'ajouter un nouveau contact avec des informations telles que le nom, le prénom, le numéro de téléphone, l'adresse e-mail, etc. Ces informations seront stockées dans un dictionnaire, et ce dictionnaire sera ajouté à une liste de contacts.
    - L'utilisateur entre une commande donnée pour accéder au service associé à cette commande
    - Le dernier dictionnaire ajouté doit être à la fin de la liste **.append**
    - La liste sera nommée **contacts**

2. Mettre à jour un contact : Vous devriez être en mesure de mettre à jour les informations d'un contact existant en recherchant le contact par son nom.
    - L'indexation nous aidera à rechercher le nom entré par l'utilisateur. Si ce nom figure dans le dictionnaire, on utilise **.pop** pour éffacer tout les éléments du dictionnaire puis on demande à l'utilisateur d'entrer les nouvelles informations.

3. Supprimer un contact : Vous devriez pouvoir supprimer un contact en recherchant le contact par son nom.
    - Utiliser l'indexation pour acéder au nom entré, tirer à partie de là, l'index du dictionnaire contenant ce contact puis utiliser **remove** pour le supprimer.

4. Rechercher un contact : Vous devriez pouvoir rechercher un contact en fonction de son nom, de son prénom ou de toute autre information disponible.
    - Utiliser l'information entré par l'utilisateur pour effectuer la recherche à la fois dans les noms, les prénoms, les numéros et les mails

5. Afficher la liste des contacts : Le programme devrait permettre d'afficher la liste complète des contacts avec toutes leurs informations.
    -utiliser tout simplement **print** pour afficher la liste

6. Utilisation de la programmation fonctionnelle : Vous pouvez utiliser des fonctions comme `filter`, `map` ou `reduce` pour effectuer des opérations sur la liste de contacts, par exemple, trouver tous les contacts avec un certain domaine d'adresse e-mail.
